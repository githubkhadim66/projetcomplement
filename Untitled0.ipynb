```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
import os, sys
from importlib import reload
from tensorflow import keras
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler


sys.path.append('..')
```

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)

print("x_train :", x_train.shape)
print("y_train :", y_train.shape)
print("x_test :", x_test.shape)
print("y_test :", y_test.shape)
```

```python
print("Before normalization : Min = {}, max = {}".format(x_train.min(),x_train.max()))

xmax = x_train.max()

x_train = x_train / xmax
x_test = x_test / xmax

print("After normalization : Min = {}, max = {}".format(x_train.min(),x_train.max()))
```

```python

# Visualisation des quelques images du dataset MNIST
plt.figure(figsize=(10, 10))
for i in range(2):
    plt.subplot(5, 5, i + 1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(x_train[i].reshape(28, 28), cmap=plt.cm.binary)
    plt.xlabel(y_train[i])
plt.show()

# Distribution des labels
# plt.figure(figsize=(8, 6))
# sns.countplot(y_train)
# plt.title("Distribution des labels dans l'ensemble d'entra√Ænement")
# plt.xlabel("Label")
# plt.ylabel("Nombre d'images")
# plt.show()
```

```python
model = keras.models.Sequential()

model.add(keras.layers.Input((28,28,1)) )
model.add(keras.layers.Conv2D(filters=8, kernel_size=(3,3), activation='relu') )
model.add(keras.layers.MaxPooling2D(pool_size=(2,2)) )
model.add(keras.layers.Dropout(0.2))

model.add(keras.layers.Conv2D(filters=16, kernel_size=(3,3), activation='relu') )
model.add(keras.layers.MaxPooling2D(pool_size=(2,2)) )
model.add(keras.layers.Dropout(0.2))

model.add(keras.layers.Flatten() )
model.add(keras.layers.Dense(100, activation='relu'))
model.add(keras.layers.Dropout(0.5))

model.add(keras.layers.Dense(10, activation='softmax'))
```

```python
model.summary()
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
from re import VERBOSE
history = model.fit(x_train, y_train,
                    batch_size=512,
                    epochs=16,
                    validation_data=(x_test, y_test),
                    verbose=1
                    )
```

```python
score = model.evaluate(x_test, y_test, verbose=1)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
```

```python
y_sigmoid = model.predict(x_test)
y_pred = np.argmax(y_sigmoid, axis=-1)

plt.figure(figsize=(10, 10))
for i in range(25):
    plt.subplot(5, 5, i + 1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(x_test[i].reshape(28, 28), cmap=plt.cm.binary)
    plt.xlabel("Pred : {}, True : {}".format(y_pred[i], y_test[i]))
plt.show()
```

```python
# prompt: represente les mauvaise prediction

import matplotlib.pyplot as plt
import numpy as np
incorrect_predictions = np.where(y_pred != y_test)[0]

plt.figure(figsize=(10, 10))
for i, idx in enumerate(incorrect_predictions[:25]):
    plt.subplot(5, 5, i + 1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(x_test[idx].reshape(28, 28), cmap=plt.cm.binary)
    plt.xlabel("Pred : {}, True : {}".format(y_pred[idx], y_test[idx]))
plt.show()
```
